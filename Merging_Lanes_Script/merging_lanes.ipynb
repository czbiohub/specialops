```python
import os
import re
import collections

fastq_pattern = re.compile(r"(.*)_L(\d+)_R(\d)_001.fastq.gz")

lanes = set()
samples = set()
mates = set()

raw_fastq_dir = "raw-fastqs"
for fname in os.listdir(raw_fastq_dir):
    matched = fastq_pattern.match(fname)
    sample_name = matched.group(1)
    lane = matched.group(2)
    mate = matched.group(3)
    samples.add(sample_name)
    lanes.add(lane)
    mates.add(mate)
```

```python
import subprocess

log_dir = "logs/"
os.makedirs(log_dir, exist_ok=True)

output_dir = "merged/"
os.makedirs(output_dir, exist_ok=True)

def merge_lanes(sample, mate):
    cmd = ["zcat"]
    for lane in ["001", "002"]:
        cmd.append(f"{raw_fastq_dir}/{sample}_L{lane}_R{mate}_001.fastq.gz")
    cmd = " ".join(cmd)
    cmd += " | gzip -c > " + os.path.join(output_dir, f"{sample}_R{mate}_001.fastq.gz")
    #print(cmd)
    with open(os.path.join(log_dir, sample), "w") as log_file:
        try:
            subprocess.run(cmd, shell=True, stderr=log_file, check=True)
        except Exception as e:
            message = (f"{sample}: job failed!", e)
            print(message)
            return message
    return f"{sample}: success!"
```

```python
import multiprocessing

with multiprocessing.Pool(90) as pool:
    job_results = pool.starmap(merge_lanes, [(sample, mate) for sample in samples for mate in [1,2]])
        
print(job_results)

with open("job_results.txt", "w") as f:
    for line in job_results:
        print(line, file=f)
```
